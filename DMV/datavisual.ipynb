```python
data['Date'] = pd.to_datetime(data['Date'])
```

```python
plt.figure(figsize=(12, 6))
plt.plot(data['Date'], data['AQI'], color='blue', marker='o')
plt.title("Overall AQI Trend Over Time")
plt.xlabel("Date")
plt.ylabel("AQI")
plt.xticks(rotation=45)
plt.grid(True)
plt.show()
```

```python
pollutants = ['PM2.5', 'PM10', 'NO', 'NO2', 'NOx', 'NH3', 'CO', 'SO2','O3','Benzene','Toluene','Xylene']  # Adjust this list as per the dataset

plt.figure(figsize=(12, 6))
for pollutant in pollutants:
    plt.plot(data['Date'], data[pollutant], label=pollutant)

plt.title("Pollutant Levels Over Time")
plt.xlabel("Date")
plt.ylabel("Pollutant Levels")
plt.legend()
plt.xticks(rotation=45)
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.boxplot(data=data[pollutants], palette='Set3')
plt.title("Distribution of Pollutant Levels")
plt.xlabel("Pollutant Type")
plt.ylabel("Pollutant Level")
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.scatterplot(data=data, x='PM2.5', y='AQI', color='red', label='PM2.5')
sns.scatterplot(data=data, x='PM10', y='AQI', color='blue', label='PM10')
plt.title("Relationship Between AQI and Pollutant Levels")
plt.xlabel("Pollutant Level")
plt.ylabel("AQI")
plt.legend()
plt.show()
```
